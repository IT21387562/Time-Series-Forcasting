```python
import tensorflow as tf
import os
import pandas as pd
import numpy as np
```

```python
zip_path = tf.keras.utils.get_file(
    origin='https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip',
    fname='jena_climate_2009_2016.csv.zip',
    extract=True)
csv_path, _ = os.path.splitext(zip_path)
```

```python
df = pd.read_csv(csv_path)
df
```

```python
df = df[5::6]  #starting from 5 take every hour
df
```

```python
df.index = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')
df[:26]
```

```python
temp = df['T (degC)']
temp.plot()
```

```python
# [[[1], [2], [3], [4], [5]]] [6]
# [[[2], [3], [4], [5], [6]]] [7]
# [[[3], [4], [5], [6], [7]]] [8]
#X - input
#y - output
def df_to_X_y(df, window_size=5):
  df_as_np = df.to_numpy()

  #[[[1], [2], [3], [4], [5]]]
  X = []

  #[6]
  y = []

  for i in range(len(df_as_np)-window_size):
    row = [[a] for a in df_as_np[i:i+window_size]]
    X.append(row)
    label = df_as_np[i+window_size]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X, y = df_to_X_y(temp, WINDOW_SIZE)
X.shape, y.shape
```

```python
#SPLIT DATA INTO TRAIN AND TEST

X_train1, y_train1 = X[:60000], y[:60000]  #1st 60000 rows
X_val1, y_val1 = X[60000:65000], y[60000:65000]
X_test1, y_test1 = X[65000:], y[65000:]

X_train1.shape, y_train1.shape, X_val1.shape, y_val1.shape, X_test1.shape, y_test1.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1 = Sequential()
model1.add(InputLayer((5, 1)))
model1.add(LSTM(64))
model1.add(Dense(8, 'relu'))
model1.add(Dense(1, 'linear')) #coz we try to predict tempeture

model1.summary()
```

```python
cp1 = ModelCheckpoint('model1/', save_best_only=True)
model1.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics=[RootMeanSquaredError()])
```

```python
model1.fit(X_train1, y_train1, validation_data=(X_val1, y_val1), epochs=10, callbacks=[cp1])
```

```python
from tensorflow.keras.models import load_model
model1 = load_model('model1/')
```

```python
train_predictions = model1.predict(X_train1).flatten()
train_results = pd.DataFrame(data={'Train Predictions':train_predictions, 'Actuals':y_train1})
train_results
```

```python
import matplotlib.pyplot as plt
plt.plot(train_results['Train Predictions'][50:100])
plt.plot(train_results['Actuals'][50:100])
```

```python
#validation results
val_predictions = model1.predict(X_val1).flatten()
val_results = pd.DataFrame(data={'Val Predictions':val_predictions, 'Actuals':y_val1})
val_results
```

```python
plt.plot(val_results['Val Predictions'][50:100])
plt.plot(val_results['Actuals'][50:100])
```

```python
test_predictions = model1.predict(X_test1).flatten()
test_results = pd.DataFrame(data={'Test Predictions':test_predictions, 'Actuals':y_test1})
test_results
```

```python
plt.plot(test_results['Test Predictions'][50:100])
plt.plot(test_results['Actuals'][50:100])
```
